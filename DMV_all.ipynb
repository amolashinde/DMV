```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.plot(df['Date'],df['AQI'],color='orange')
plt.title("Date vs AQI")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```python
pollutants = ['PM2.5','PM10','CO']
for p in pollutants:
    plt.plot(df['Date'],df[p],label=p)
plt.title("Date vs Pollutants")
plt.xlabel("Date")
plt.ylabel("Concentration")
plt.legend()
plt.show()
```

```python
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['AQI'].mean()
monthly_avg.plot(kind='bar',color='teal')
plt.title("Bar plot of AQI by time period")
plt.xlabel('Month')
plt.ylabel('AQI')
plt.show()
```

```python
df = pd.read_csv("3_tele_com.csv")
```

```python
df.isnull().sum()
```

```python
df['MultipleLines'] = df['MultipleLines'].fillna('Not known')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df['InternetService'] = df['InternetService'].replace("Fiber opticalal","Fiber optic")
```

```python
df
```

```python
df.dtypes
```

```python
df['MonthlyCharges'] = df['MonthlyCharges'].astype(int)
df['TotalCharges'] = df['TotalCharges'].astype(int)
```

```python
df.dtypes
```

```python
df['Contract_renew'] = df['Contract'].apply(lambda x:"yes" if x in ['One year', 'Two year'] else "No")
```

```python
df['Contract_renew']
```

```python
import numpy as np

numric_data = ['tenure', 'MonthlyCharges', 'TotalCharges']
z_score = np.abs((df[numric_data] - df[numric_data].mean()) / df[numric_data].std())
threshold = 3
outlier = z_score > threshold
print(outlier)

outlier_count = outliers.sum()
print(outlier_count)
```

```python
from sklearn.model_selection import train_test_split
```
